```python
import os
from deepface import DeepFace
from tqdm import tqdm
import pandas as pd
import cv2
# import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
```

```python
target_img_name="CommonPeople/photo_4.jpg"
```

```python
characters=[]
for dirPath, dirnames, filenames in os.walk("CelebrityPeople/"):
    for filename in filenames:
        if ".jpg" in filename:
            characters.append(dirPath + filename)
```

```python
characters
```

```python
similarties ={}
for character in tqdm(characters):
    obj=DeepFace.verify(img1_path=character, img2_path=target_img_name,
                        model_name="Facenet",detector_backend="mtcnn",distance_metric="euclidean")
    similarties[character]=obj["distance"]
    
```

```python
similarties
```

```python
df=pd.DataFrame(similarties.items(),columns=["character","distance"])
```

```python
df
```

```python
df=df.sort_values(by=["distance"])
```

```python
df
```

```python
imgname=df.iloc[0]["character"]
# img=cv2.imread(imgname)


img =DeepFace.detectFace(img_path=imgname,detector_backend="mtcnn")
```

```python
targetimg =DeepFace.detectFace(img_path=target_img_name,detector_backend="mtcnn")
```
